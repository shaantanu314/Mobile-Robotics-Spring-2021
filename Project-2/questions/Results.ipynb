```python
import cv2
from scipy import ndimage
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('obh.jpg')
# print(img.shape)
crop_img = img[50:50+300, 60:60+600]
result = cv2.resize(crop_img,(700,500))
# print('rezised')
result = ndimage.rotate(result, 20)
cv2.imwrite('obh_rz.jpg',result)
```

```python

img1 = cv2.imread('obh.jpg')  
img2 = cv2.imread('obh_r.jpg')
sift = cv2.xfeatures2d.SIFT_create()

kp1, descriptors_1 = sift.detectAndCompute(img1,None)
kp2, descriptors_2 = sift.detectAndCompute(img2,None)


bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)

matches = bf.match(descriptors_1,descriptors_2)
matches = sorted(matches, key = lambda x:x.distance)

print(descriptors_1.shape)

img3 = cv2.drawMatches(img1, kp1, img2, kp2, matches[:50], img2, flags=2)

cv2.imwrite('fuck.jpg',img3)
plt.figure(figsize=(15,20))
plt.imshow(img3)
plt.title('Rotation Only')

    
```
